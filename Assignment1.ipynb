# Using Function

#Plot y = g(f(x)) where u = f(x) = wx + b and y = g(u) = 1/ (1 + e^(-u))

```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions
def f(x, w, b):
    return w * x + b

def g(u):
    return 1 / (1 + np.exp(-u))

# Parameters
w = 1
b = 0

# Values of x
x = np.arange(-10, 10, 0.1)

# Calculate y values
u = f(x, w, b)
y = g(u)

# Plot
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = g(f(x)) where f(x) = wx + b and g(u) = 1 / (1 + e^(-u))')
plt.grid(True)
plt.show()
```

#Plot y = g(f(x)) where f(x) = wx + b, u = f(x), and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters Weight and Bias
w = 5
b = 10

# Values of x
x = np.arange(-10, 10, 0.1)

# Calculate f(x) = wx + b
u = w * x + b

# Calculate g(u) = (e^u - e^(-u)) / (e^u + e^(-u))
y = (np.exp(u) - np.exp(-u)) / (np.exp(u) + np.exp(-u))

# Plot
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = g(f(x)) where f(x) = wx + b and g(u) = (e^u - e^(-u)) / (e^u + e^(-u))')
plt.grid(True)
plt.show()
```

#Plot y1 = g1(f1(x)) where f1(x) = w1x + b1, u1 = f1(x), and g1(u1) = 1/ (1 + e^(-u1))


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
w1 = 1
b1 = 0

# Values of x
x = np.arange(-10, 10, 0.1)

# Calculate f1(x) = w1x + b1
u1 = w1 * x + b1

# Calculate g1(u1) = 1 / (1 + e^(-u1))
y1 = 1 / (1 + np.exp(-u1))

# Plot
plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('y1')
plt.title('Plot of y1 = g1(f1(x)) where f1(x) = w1x + b1 and g1(u1) = 1 / (1 + e^(-u1))')
plt.grid(True)
plt.show()
```

#Plot y2 = g2(f2(x)) where f2(x) = w2x + b2, u2 = f2(x), and g2(u2) = 1/ (1 + e^(-u2))


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
w2 = 5
b2 = 10

# Values of x
x = np.arange(-10, 10, 0.1)

# Calculate f2(x) = w2x + b2
u2 = w2 * x + b2

# Calculate g2(u2) = 1 / (1 + e^(-u2))
y2 = 1 / (1 + np.exp(-u2))

# Plot
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y2')
plt.title('Plot of y2 = g2(f2(x)) where f2(x) = w2x + b2 and g2(u2) = 1 / (1 + e^(-u2))')
plt.grid(True)
plt.show()
```

#Plot y = g3(f3(v)) where f3(v) = w3y1 + w4y2 + b, w = f3(v), and g3(w) = 1/ (1 + e^(-w))

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
w3 = 5
w4 = 5
b = 0

# Values of y1 and y2 (output from previous functions)
y1 = 1 / (1 + np.exp(-u1))  # assuming u1 is already calculated from previous computation
y2 = 1 / (1 + np.exp(-u2))  # assuming u2 is already calculated from previous computation

# Calculate f3(v) = w3*y1 + w4*y2 + b
v = w3 * y1 + w4 * y2 + b

# Calculate g3(w) = 1 / (1 + e^(-w))
y = 1 / (1 + np.exp(-v))

# Plot
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = g3(f3(v)) where f3(v) = w3*y1 + w4*y2 + b and g3(w) = 1 / (1 + e^(-w))')
plt.grid(True)
plt.show()
```
