```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
a=3 
b=10

x=np.arange(-10,11)
y = a * x + b
plt.plot(x,y)
plt.xlim(-10,10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = ax+b')
plt.show()
```

```python
a=3 
b=10
step_size=0.01
x=np.arange(-10,11,0.01)
sin=np.sin(x)
y = a * sin + b
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10,10)
plt.title('Plot of y = asin(x)+b')
plt.show()
```

```python
a=3 
b=10
step_size=0.01
x=np.arange(-10,10.01,0.01)
y = a * np.exp(x) + b
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-10,10)
plt.title('Plot of y = ae^x+b')
plt.show()
```
